```python
df['month'] = pd.to_datetime(df['published']).dt.month

vuln_counts = df.groupby('month').size()

plt.figure(figsize=(12,6))
plt.plot(vuln_counts.index, vuln_counts.values)
plt.title('seasonality ?')
plt.xlabel('Month')
plt.ylabel('Vulnerabilities')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['published'] = pd.to_datetime(df['published'])
vuln_counts = df.resample('Y', on='published').size()

plt.figure(figsize=(12,6))
plt.plot(vuln_counts.index, vuln_counts.values)
plt.xlabel('Publication Date')
plt.ylabel('Number of Vulnerabilities')
plt.title('Yearly Vulnerabilities')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df['published'] = pd.to_datetime(df['published'])

# Create separate counts for each severity
severity_counts = df.groupby([pd.Grouper(key='published', freq='M'), 'metrics_severity']).size().unstack(fill_value=0)

# Count automatable vulnerabilities
df['Automatable'] = df['Automatable'].map({'Yes': 1, 'No': 0})
auto_counts = df.groupby(pd.Grouper(key='published', freq='M'))['Automatable'].sum()

fig, ax1 = plt.subplots(figsize=(15,7))

# Plot severities as stacked areas
severity_counts.plot(kind='area', ax=ax1, alpha=0.6)

# Plot automatable line on secondary y-axis
ax2 = ax1.twinx()
ax2.plot(auto_counts.index, auto_counts.values, color='red', linewidth=2, label='Automatable')

ax1.set_xlabel('Publication Date')
ax1.set_ylabel('Number of Vulnerabilities by Severity')
ax2.set_ylabel('Number of Automatable Vulnerabilities')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Vulnerabilities Over Time by Severity and Automatability')
plt.grid(True)
plt.tight_layout()
plt.show()
```
